```python
import gps_walkthrough as gpswt 
import numpy as np
```

```python
res = []
for prn in range(1, 38):
    ca_gen = gpswt.GpsCaCode(prn_id=prn, sampling_dt=1/4.096e6).samples
    ca_gen_twos = [0x7fff if x == 1 else 0x8000 for x in ca_gen]
    res.extend(ca_gen_twos)
```

```python
hex_res = [f"{int(x):04x}" for x in res]
```

```python
hex_res
```

```python
len(hex_res)/4096
```

```python
def decode(x):
    return 1 if x == '7fff' else -1
a = 1
b = 4
N = 4096
sum(decode(x)*decode(y) for x, y in zip(hex_res[(a-1)*N:a*N], hex_res[(b-1)*N:b*N]))
```

```python
with open('ca.mem', 'w') as f:
    for entry in hex_res:
        for char in entry:
            print(char, file=f)
```
