```python
import gps_walkthrough as gpswt 
import numpy as np
```

```python
res = []
for prn in range(1, 38):
    ca_gen = gpswt.GpsCaCode(prn_id=prn, sampling_dt=1/6.4e6).samples
    res = np.hstack([res, (ca_gen+1)/2 * ((1<<15)-1), [0] * (8192-6400)])
```

```python
hex_res = [f"{int(x):04x}" for x in res]
```

```python
hex_res
```

```python
sum(int(x) for x in hex_res[8*8192+6400:9*8192])
```

```python
with open('ca.mem', 'w') as f:
    for entry in hex_res:
        for char in entry:
            print(char, file=f)
```
