```python
import numpy as np
from pynq import allocate
```

```python
from pynq import PL
PL.reset() #important fixes caching issues which have popped up.
from pynq import Overlay  #import the overlay module
ol = Overlay('./design_1.bit')  #locate/point to the bit file
```

```python
mmio_interface = ol.mmio_interface_0
dma = ol.axi_dma_0
```

```python
mmio_interface.write(0x04, 5) # request prn ID 6
```

```python
out_buffer = allocate(4096, dtype=np.int32)
dma.recvchannel.transfer(out_buffer)
dma.recvchannel.wait()
```
